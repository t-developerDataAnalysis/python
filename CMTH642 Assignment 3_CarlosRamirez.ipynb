### Question 6
Reduce the levels of rating for quality to two levels, Pass and Fail. Assign the levels 3, 4, and 5 to level Fail; and 6, 7, 8, and 9 to level Pass. (10
points)

```python
# INSERT YOUR ANSWER HERE
#create a function 
# define a function to map the original levels to the new levels
def map_levels(x):
    if x in [3, 4, 5]:
        return 0
    elif x in [6, 7, 8, 9]:
        return 1
    else:
        return None  # or return x if you want to keep other values as they are

# apply the mapping function to the column and create a new column with the new levels
wine_data_all['level'] = wine_data_all['quality'].apply(map_levels)

# print the new DataFrame
wine_data_all.head()
```

### Question 7
Normalize the numeric attributes. Hint: (x - min(x)) / (max(x) - min(x)) (10 points)

```python
# INSERT YOUR ANSWER HERE
wine_data_all_normalized = wine_data_all.loc[:,'fixed acidity':'quality'].apply(lambda x: (x - x.min()) / (x.max() - x.min()))

wine_data_all_normalized['level'] = wine_data_all['level'].copy()

wine_data_all_normalized.head(10)


```

### Question 8
Divide the dataset to training and test sets. (10 points)

```python
# INSERT YOUR ANSWER HERE
#split dataset in features and target variable
feature_cols = ["fixed acidity","volatile acidity","citric acid","residual sugar","chlorides",
                "free sulfur dioxide","total sulfur dioxide","density","pH","sulphates","alcohol"]
X = wine_data_all_normalized[feature_cols] # Features
y = wine_data_all_normalized.level # Target variable

#Splitting the data
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.25, random_state=16)
```

### Question 9
Use the Logistic Regression algorithm to predict the quality of wine using its attributes. (12 points)

```python
# INSERT YOUR ANSWER HERE
# Logistic Regression

#Model Development and prediction
# instantiate the model (using the default parameters)
from sklearn.linear_model import LogisticRegression

logreg = LogisticRegression(random_state=16)

# fit the model with data
logreg.fit(X_train, y_train)

y_pred_logistic = logreg.predict(X_test)

#showing the results
from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred_logistic))

from sklearn.metrics import accuracy_score, precision_score, recall_score


#accuracy
accuracy_pred_knn = accuracy_score(y_test, y_pred_logistic)
print("Accuracy:", accuracy_pred_knn)

#presicion
precision_pred_knn = precision_score(y_test, y_pred_logistic)
print("Precision: ", precision_pred_knn )

#recall score
recall_score_knn = recall_score(y_test, y_pred_logistic)
print("recall score ", recall_score_knn )



```

### Question 10
Use the KNN algorithm to predict the quality of wine using its attributes. (12 points)

```python
# INSERT YOUR ANSWER HERE
from sklearn.neighbors import KNeighborsClassifier
knn = KNeighborsClassifier(n_neighbors=3)
knn.fit(X_train, y_train)

y_pred_knn = knn.predict(X_test)

from sklearn.metrics import accuracy_score, precision_score, recall_score

from sklearn.metrics import classification_report
print(classification_report(y_test,y_pred_knn))

#accuracy
accuracy_pred_knn = accuracy_score(y_test, y_pred_knn)
print("Accuracy:", accuracy_pred_knn)

#presicion
precision_pred_knn = precision_score(y_test, y_pred_knn)
print("Precision: ", precision_pred_knn )

#recall score
recall_score_knn = recall_score(y_test, y_pred_knn)
print("recall score ", recall_score_knn )
```
